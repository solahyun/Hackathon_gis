```python
data=pd.read_csv('/content/202001~202109.csv',encoding='euc-kr')
data.set_index(['지역','연월'],inplace=True)
data
```

```python
col=[]
for i in data.columns:
  col.append(i)

area = ['동구','서구','남구','북구','광산구']
agri = col[:8]
anim = col[8:13]
fish = col[13:20]
fruit = col[20:23]
vegi = col[23:31]
life = col[31:43]
oil = col[43:48]
bath = col[48:52]
```

```python
# j = 0,1,2,3,4 >> 동,서,남,북,광산구
j=4
agri_data = [np.round((pd.DataFrame(data.loc[area[j]][agri[i]])[agri[i]][-1])/
                 (pd.DataFrame(data.loc[area[j]][agri[i]])[agri[i]][-2])-1,3)
        for i in range(len(agri))]
anim_data = [np.round((pd.DataFrame(data.loc[area[j]][anim[i]])[anim[i]][-1])/
                 (pd.DataFrame(data.loc[area[j]][anim[i]])[anim[i]][-2])-1,3)
        for i in range(len(anim))]
fruit_data = [np.round((pd.DataFrame(data.loc[area[j]][fruit[i]])[fruit[i]][-1])/
                 (pd.DataFrame(data.loc[area[j]][fruit[i]])[fruit[i]][-2])-1,3)
        for i in range(len(fruit))]
vegi_data = [np.round((pd.DataFrame(data.loc[area[j]][vegi[i]])[vegi[i]][-1])/
                 (pd.DataFrame(data.loc[area[j]][vegi[i]])[vegi[i]][-2])-1,3)
        for i in range(len(vegi))]
life_data = [np.round((pd.DataFrame(data.loc[area[j]][life[i]])[life[i]][-1])/
                 (pd.DataFrame(data.loc[area[j]][life[i]])[life[i]][-2])-1,3)
        for i in range(len(life))]
oil_data = [np.round((pd.DataFrame(data.loc[area[j]][oil[i]])[oil[i]][-1])/
                 (pd.DataFrame(data.loc[area[j]][oil[i]])[oil[i]][-2])-1,3)
        for i in range(len(oil))]
bath_data = [np.round((pd.DataFrame(data.loc[area[j]][bath[i]])[bath[i]][-1])/
                 (pd.DataFrame(data.loc[area[j]][bath[i]])[bath[i]][-2])-1,3)
        for i in range(len(bath))]
df_agri = pd.DataFrame([agri_data],index=['전월대비'],columns=[agri]).transpose()
df_anim = pd.DataFrame([anim_data],index=['전월대비'],columns=[anim]).transpose()
df_fruit = pd.DataFrame([fruit_data],index=['전월대비'],columns=[fruit]).transpose()
df_vegi = pd.DataFrame([vegi_data],index=['전월대비'],columns=[vegi]).transpose()
df_life = pd.DataFrame([life_data],index=['전월대비'],columns=[life]).transpose()
df_oil = pd.DataFrame([oil_data],index=['전월대비'],columns=[oil]).transpose()
df_bath = pd.DataFrame([bath_data],index=['전월대비'],columns=[bath]).transpose()
df_max = pd.DataFrame(data=[agri_data[np.argmax(np.abs(agri_data))],anim_data[np.argmax(np.abs(anim_data))],fruit_data[np.argmax(np.abs(fruit_data))],vegi_data[np.argmax(np.abs(vegi_data))],life_data[np.argmax(np.abs(life_data))],oil_data[np.argmax(np.abs(oil_data))],bath_data[np.argmax(np.abs(bath_data))]],
                       columns=['전월대비'],
                       index=[agri[np.argmax(np.abs(agri_data))],anim[np.argmax(np.abs(anim_data))],fruit[np.argmax(np.abs(fruit_data))],vegi[np.argmax(np.abs(vegi_data))],life[np.argmax(np.abs(life_data))],oil[np.argmax(np.abs(oil_data))],bath[np.argmax(np.abs(bath_data))]])
df_max
```
